```python
import csv
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://ocw.mit.edu/courses/1-00-introduction-to-computers-and-engineering-problem-solving-spring-2012/pages/lecture-notes/')
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print( soup.prettify() )
```

```python
headers=soup.find_all('th')
headers=[header.get_text().strip() for header in headers]
columns=[]
for i in headers:
    lis=list(i)
    if "Â" in lis:
        lis.remove("Â")
        lis="".join(lis)
    else:
        lis="".join(lis)
    columns.append(lis)

print(columns)
```

```python
columns.append("TOPICS_URL")
columns.append("FILES_URL")
columns
```

```python
df=pd.DataFrame([],columns=['LEC\xa0#', 'TOPICS', 'TOPICS_URL', 'SUPPORTING\xa0FILES','FILES_URL'])
df
```

```python
rows=soup.find_all("tr")[1:]
len(rows)
```

```python
for idx,row in enumerate(rows): 
    lis=row.text.strip().split("\n")
    for i in range(len(lis)):
        if "" in lis:
            lis.remove("")
        elif "Â" in lis:
            lis.remove("Â")
    
    if len(row.find_all("a"))==0:
         topic_url=""
         files_url=""
    elif len(row.find_all("a"))==1:
         topic_url="https://ocw.mit.edu"+row.find_all("a")[0].get("href")
         files_url=""
    else:
         topic_url="https://ocw.mit.edu"+row.find_all("a")[0].get("href")
         files_url="https://ocw.mit.edu"+row.find_all("a")[1].get("href")

    if len(lis)==1:
        df.loc[idx,"TOPICS"]=lis[0]
    elif len(lis)==2:
        df.loc[idx,"LEC\xa0#"]=lis[0]      
        df.loc[idx,"TOPICS"]=lis[1]
    elif len(lis)==3:
        df.loc[idx,"LEC\xa0#"]=lis[0]      
        df.loc[idx,"TOPICS"]=lis[1]
        df.loc[idx,"SUPPORTING\xa0FILES"]=lis[2]

    df.loc[idx,"TOPICS_URL"]=topic_url
    df.loc[idx,"FILES_URL"]=files_url


    print(topic_url)
    print(files_url)
    print()
```

```python
df["LEC\xa0#"].replace(np.nan,"",inplace=True)
df["TOPICS"].replace(np.nan,"",inplace=True)
df["SUPPORTING\xa0FILES"].replace(np.nan,"",inplace=True)
```
